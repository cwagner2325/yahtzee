```python
from ipywidgets import Textarea, Button, VBox, HBox, Layout, ButtonStyle, Text
from random import *

global lockInChoice
global total
global runningUpperSum

strings = ['Three of a Kind', 'Four of a Kind', 'Small Straight', 'Large Straight', 'Full House', 'Yahtzee', 'Bonus']
rollCounter = 3
lowerHalf = []
upperHalf = []
diceButtons = []
lockInButtons = []
texts = []
rows =[]
dice = [5] 

lockedIn = {0 : False, 1 : False, 2 : False, 3 : False, 4 : False, 5 : False, 6 : False, 7 : False, 8 : False, 9 : False,
            10 : False, 11 : False, 12 : False}

lockInChoice = None
total = 0
runningUpperSum = 0

def resetDice():
    global dice
    dice.clear()
    newRoll = 0
    
    for i in range(5):
        if diceButtons[i].style.button_color == 'lightgreen':
            diceButtons[i].style.button_color = 'lightgray'
        
        diceButtons[i].description = str(int(newRoll))
        dice.append(str(int(newRoll)))
                
def rollDice(item = None):
    global dice
    dice.clear()
    if rollCounter > 0:
        for i in range(5):
            newRoll = (random()*100//1%6+1)
            dice.append(diceButtons[i].description)
            if diceButtons[i].style.button_color != 'lightgreen':
                diceButtons[i].description = str(int(newRoll))
                dice[i] = str(int(newRoll))
    
        decrementRollCounter()
        calculateScores(dice)
        
def calculateScores(dice):
    counts = []
    sumDice = 0
    for dices in dice:
        sumDice = sumDice + int(dices)
    counts.append(dice.count('1'))
    counts.append(dice.count('2'))
    counts.append(dice.count('3'))
    counts.append(dice.count('4'))
    counts.append(dice.count('5'))
    counts.append(dice.count('6'))
    
    if(not(lockedIn[0])):
        texts[0].value = str(counts[0])
    if(not(lockedIn[1])):
        texts[1].value = str(2*counts[1])
    if(not(lockedIn[2])):
        texts[2].value = str(3*counts[2])
    if(not(lockedIn[3])):
        texts[3].value = str(4*counts[3])
    if(not(lockedIn[4])):
        texts[4].value = str(5*counts[4])
    if(not(lockedIn[5])):
        texts[5].value = str(6*counts[5])
    
    if(not(lockedIn[6])):
        if(3 in counts or 4 in counts or 5 in counts):
            texts[6].value = str(sumDice)
        else:
            texts[6].value = '0'
     
    if(not(lockedIn[7])):
        if(4 in counts or 5 in counts):
            texts[7].value = str(sumDice)
        else:
            texts[7].value = '0'
   
    if(not(lockedIn[8])):
        if((counts[0] >= 1 and counts[1] >= 1 and counts[2] >= 1 and counts[3] >= 1) or
           (counts[1] >= 1 and counts[2] >= 1 and counts[3] >= 1 and counts[4] >= 1) or 
            counts[2] >= 1 and counts[3] >= 1 and counts[4] >= 1 and counts[5] >= 1):
            texts[8].value = '30'
        else:
            texts[8].value = '0'
        
    if(not(lockedIn[9])):
        if((counts[0] >= 1 and counts[1] >= 1 and counts[2] >= 1 and counts[3] >= 1 and counts[4] >= 1) or
            counts[1] >= 1 and counts[2] >= 1 and counts[3] >= 1 and counts[4] >= 1 and counts[5] >= 1):
            texts[9].value = '40'
        else:
            texts[9].value = '0'
     
    if(not(lockedIn[10])):
        if(2 in counts and 3 in counts):
            texts[10].value = '25'
        else:
            texts[10].value = '0'
    
    if(not(lockedIn[11])):
        if(5 in counts):
            texts[11].value = '50'
        else:
            texts[11].value = '0'
    
    if(not(lockedIn[12])):
        texts[12].value = str(sumDice)
    
def lockIn(item):
    if item.description != '0':
        if item.style.button_color != 'lightgreen':
            item.style.button_color = 'lightgreen'
        else:
            item.style.button_color = 'lightgray'

def lockInCategory(item):
    global lockInChoice
        
    if None == lockInChoice:
        lockInChoice = item
        item.style.button_color = 'lightgreen'
    else:
        if item == lockInChoice:
            if not lockedIn[lockInButtons.index(lockInChoice)]:
                item.style.button_color = 'lightgray'
            lockInChoice = None
        elif not lockedIn[lockInButtons.index(lockInChoice)]:
            lockInChoice.style.button_color = 'lightgray'
            lockInChoice = item
            item.style.button_color = 'lightgreen'
        else:
            lockInChoice = item
            item.style.button_color = 'lightgreen'
            
def decrementRollCounter():
    global rollCounter
    rollCounter = rollCounter - 1
    rollCount.value = str(rollCounter)
    
def incrementRollCounter():
    global rollCounter
    rollCounter = 3
    rollCount.value = str(rollCounter)
    
def updateFinal():
    global runningUpperSum
    global total
    if runningUpperSum >= 63:
        total = total + 35

    finalTotal.value=f'Your Final Score is {total}'
    
    
def initNextTurn(item = None):
    global lockInChoice
    global total
    global runningUpperSum
    
    if(None == lockInChoice):
        pass
    elif (lockedIn[lockInButtons.index(lockInChoice)] == True):
        pass
    else:
        incrementRollCounter()
        lockInChoice.description = 'Locked In!'
        lockedIn[lockInButtons.index(lockInChoice)] = True
        if lockInButtons.index(lockInChoice) < 6:
            runningUpperSum = runningUpperSum + int(texts[lockInButtons.index(lockInChoice)].value)
            upperSum.value = str(runningUpperSum)
        total = total + int(texts[lockInButtons.index(lockInChoice)].value)
        runningTotal.value = str(total)
        lockInChoice = None
        resetDice()
        if (False not in lockedIn.values()):
            updateFinal()
        
for i in range (1, 7, 1):
    button = Button(description='Lock It In?', layout=Layout(width='auto', height='auto', border='1px solid black'), 
                   style=ButtonStyle(button_color = 'lightgray'))
    text = Textarea(value=str(0), rows=1, description=(f"{i}'s"), 
                    layout=Layout(height='auto', width="auto"), disabled=True)
    button.on_click(lockInCategory)
    lockInButtons.append(button)
    texts.append(text)
    row = HBox([text, button])
    upperHalf.append(row)
    
for i in range(7):
    button = Button(description='Lock It In?', layout=Layout(width='auto', height='auto', border='1px solid black'), 
                   style=ButtonStyle(button_color = 'lightgray'))
    text = Textarea(value='0', rows=1, description=strings[i],
                    layout=Layout(height='100%', width="20%"), disabled=True)
    button.on_click(lockInCategory)
    lockInButtons.append(button)
    texts.append(text)
    row = HBox([text, button])
    lowerHalf.append(row)
    
for i in range(5):
    diceButton = Button(description='0', layout=Layout(width='auto', height='25px', border='1px solid black'), 
                       style=ButtonStyle(button_color='lightgray'))
    diceButtons.append(diceButton)
    diceButton.on_click(lockIn)
    
rollButton = Button(description='Roll!', layout=Layout(width='97%', height='25px', border='1px solid black'))
rollCount = Textarea(value=str(rollCounter), rows=1, description='Rolls Remaining: ',
                    layout=Layout(height='auto', width='45px'), disabled=True)
rollButton.on_click(rollDice)

nextTurn = Button(description='Next Turn!', layout=Layout(width='60%', height='25px', border='1px solid black'))
runningTotal = Textarea(description='Total: ', value='0', rows=1,
                        layout=Layout(height='25px', width='20%'), disabled=True)
upperSum = Textarea(description='Upper Sum: ', value='0', rows=1,
                        layout=Layout(height='25px', width='40%'), disabled=True)
finalTotal = Textarea(description='Final Total: ', value='', rows=1,
                        layout=Layout(height='25px', width='400px'), disabled=True)
nextTurn.on_click(initNextTurn)

diceButtons2 = VBox([HBox(diceButtons), rollButton, VBox([rollCount])])
diceButtons2.layout.border = '3px solid black'

lowerHalf2 = VBox(lowerHalf)
runningTotal2 = VBox([runningTotal])
upperSum2 = HBox([upperSum])
upperSum2.layout.border = '1px solid black'
runningTotal2.layout.border = '1px solid black'
upperHalf2 = HBox([VBox(upperHalf), diceButtons2])

VBox([upperHalf2, upperSum2, lowerHalf2, runningTotal2, VBox([nextTurn]), finalTotal])
```
